```python
div      = np.load('data/div.npy')       #depth-averaged horizontal divergence
FCOR     = np.load('data/FCOR.npy')      #Coriolis parameter
grad_B   = np.load('data/grad_B.npy')    #depth-averaged buoyancy gradient magnitude
HBL      = np.load('data/HBL.npy')       #boundary layer depth
HML      = np.load('data/HML.npy')       #mixed-layer depth
Nsquared = np.load('data/Nsquared.npy')  #Brunt-Vaisala frequency
Q        = np.load('data/Q.npy')         #surface heat flux
strain   = np.load('data/strain.npy')    #depth averaged strain magnitude
TAU      = np.load('data/TAU.npy')       #surface wind stress magnitude
vort     = np.load('data/vort.npy')      #depth-averaged vertical vorticity
WB_sg    = np.load('data/WB_sg.npy')     #depth-averaged vertical buoyancy flux
```

```python
plt.figure(figsize=(25,10))

plt.subplot(2,5,1)
plt.contourf(grad_B[2,:,:])
plt.title('depth-averaged buoyancy gradient magnitude \n $|\overline{\mathcal{b}}^\mathcal{z}|$')

plt.subplot(2,5,2)
plt.contourf(FCOR[2,:,:])
plt.title('Coriolis parameter \n $\overline{\mathcal{f}}$')

plt.subplot(2,5,3)
plt.contourf(HML[2,:,:])
plt.title('mixed-layer depth \n $\overline{H_{ML}}$')

plt.subplot(2,5,4)
plt.contourf(Nsquared[2,:,:])
plt.title('Brunt-Visl frequency \n $\overline{N^2}$')

plt.subplot(2,5,5)
plt.contourf(Q[2,:,:])
plt.title('surface heat flux \n $\overline{Q*}$')

plt.subplot(2,5,6)
plt.contourf(TAU[2,:,:])
plt.title('surface wind stress magnitude \n $\overline{||} = \sqrt{\overline{_\mathcal{x}}^2 + \overline{_\mathcal{y}}^2}$')

plt.subplot(2,5,7)
plt.contourf(HBL[2,:,:])
plt.title('boundary layer depth \n $H_BL$')

plt.subplot(2,5,8)
plt.contourf(strain[2,:,:])
plt.title('depth-averaged strain magnitude \n $\overline{\sigma}^2 = \sqrt{(\overline{\mathcal{u}_\mathcal{x}}^\mathcal{z} - \overline{\mathcal{v}_\mathcal{y}}^\mathcal{z})^2 + (\overline{\mathcal{v}_\mathcal{x}}^\mathcal{z} + \overline{\mathcal{u}_\mathcal{y}}^\mathcal{z})^2}$ ')

plt.subplot(2,5,9)
plt.contourf(vort[2,:,:])
plt.title('depth-averaged vertical vorticity \n $\overline{\zeta}^\mathcal{z} = \overline{\mathcal{v}_\mathcal{x}}^\mathcal{z} - \overline{\mathcal{u}_\mathcal{y}}^\mathcal{z}$')

plt.subplot(2,5,10)
plt.contourf(div[2,:,:])
plt.title('depth-averaged horizontal divergence \n $\overline{\delta^\mathcal{z}} = \overline{\mathcal{u}_\mathcal{x}}^z + \overline{\mathcal{v}_\mathcal{y}}^\mathcal{z}$')


plt.subplots_adjust(hspace = 0.3, wspace=0.25)
```
